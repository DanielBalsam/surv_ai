```python
from core.tools.toolbelt import Toolbelt
from core.tools.research.new_york_times import NewYorkTimesTool
from core.tools.research.wikipedia import WikipediaTool
from lib.agent_log import agent_log, AgentLogLevel
from core.assembly.voting_research import VotingResearcherAssembly

agent_log.set_log_level(AgentLogLevel.SPEECH)
date_ranges = [
    ('20211101', '20220201'),
    ('20220201', '20220501'),
    ('20220501', '20220801'),
    ('20220801', '20221101'),
    ('20221101', '20230201'),
    ('20230201', '20230501'),
]

results = {}

for range in date_ranges:
    assembly = VotingResearcherAssembly(client, Toolbelt(client, [NewYorkTimesTool(client, n_pages=5, begin_date=range[0], end_date=range[1])]),n_agents=30)
    results[range] =  await assembly.prompt("The US Economy will experience a recession in the next year.")

results 
```

```python
from core.assembly.interfaces import AssemblyResponse

cummulative = {}

total_in_favor = 0
total_against = 0

for key, val in results.items():
    total_in_favor += val.in_favor
    total_against += val.against

    cummulative[key] = AssemblyResponse(in_favor=total_in_favor, against=total_against, undecided=0, percent_in_favor=total_in_favor/(total_in_favor + total_against), error_bar=0, summaries=[])

cummulative
```

```python
import matplotlib.pyplot as plt
  
# plotting the points 
plt.plot([r[0] for r in results.keys()], [r.percent_in_favor for r in results.values()])
  
# naming the x axis
plt.xlabel('Date')
# naming the y axis
plt.ylabel('Recession sentiment')
  
# giving a title to my graph
plt.title('Recession sentiment of New York Times by quarter')
  
# function to show the plot
plt.show()
```
